```python
import pandas as pd
import requests
import json
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
url = 'https://finance.yahoo.com/quote/BTC-USD/history/?period1=1410912000&period2=1713963810'
agent = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/124.0.0.0 Safari/537.36'}
response = requests.get(url, headers= agent)
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
items = soup.find_all('tr', attrs= {'class': 'svelte-ewueuo'})
(items[1].get_text())
```

```python
sequence = "Apr 24, 2024 66,420.40 67,053.77 64,529.95 64,529.95 64,529.95 25,746,866,176"
ite = sequence.split()
# Extract the date part and convert it to the desired format
date_str = ite[0] + " " + ite[1] + " " + ite[2]
date_obj = datetime.strptime(date_str, "%b %d, %Y")
formatted_date = date_obj.strftime("%Y/%m/%d")
```

```python
formatted_date
```

```python
float(items[2].get_text().split(' ')[6].replace(',', ''))
```

```python
key = 0
df = {}
for i in items:
    row = i.get_text().split(' ')
    
    if items.index(i) == 3670:
        break
    elif items.index(i) == 0:
        continue
    else:
        df[key] = {
            'day': row[1].strip(','),
            'month': row[0],
            'year': row[2],
            'open': float(row[3].replace(',', '')),
            'high': float(row[4].replace(',', '')),
            'low': float(row[5].replace(',', '')),
            'close': float(row[6].replace(',', '')),
        }
        key+= 1
df
```

```python
bitcoin = pd.DataFrame(df).T
bitcoin
```

```python
def format_date(row):
    date_str = row['year'] + ' ' + row['month'] + ' ' + row['day']
    date_obj = datetime.strptime(date_str, "%Y %b %d")
    formatted_date = date_obj.strftime('%Y/%m/%d')
    return formatted_date
bitcoin['date'] = bitcoin.apply(format_date, axis=1)
bitcoin.drop(columns= ['day', 'month', 'year'], inplace=True)
```

```python
bitcoin = bitcoin.iloc[:,[4,0,1,2,3]]
bitcoin
```

```python
bitcoin.to_csv('Bitcoin.csv')
```
